```python
import requests
```

```python
url ='https://www.flipkart.com/search?q=bags&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off' 
response = requests.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup= BeautifulSoup(response.text)
soup.text
```

```python
soup.title
```

```python
soup.find('a')
```

```python
soup.a
```

```python
links = soup.findAll('a')
for link in links:
    print(link.text)
    print(link.get('href'))
```

```python
links = soup.findAll('a')
for link in links:
    print(link.text)
    print(link.get('h1'))
```

```python
headings=soup.findAll('h3')
print('total headings found:',len(headings))
for item in headings:
    print(item.text)
```

```python
area = soup.find_all('div', class_ = '_1YokD2 _3Mn1Gg')
print(len(area))
```
